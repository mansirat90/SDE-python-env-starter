## Section 1.1: Correlation Matrices (Guided)

### Introduction to Correlation Matrices

Let's get a better understanding of our data by using a **correlation matrix**.

A correlation matrix is a table showing correlation coefficients between many variables. Each cell in the table displays the correlation between two variables.

**How to interpret correlation values:**

| Correlation Range | Interpretation |
|---|---|
| .90 to 1.00 (-.90 to -1.00) | Very high positive (negative) correlation |
| .70 to .90 (-.70 to -.90) | High positive (negative) correlation |
| .50 to .70 (-.50 to -.70) | Moderate positive (negative) correlation |
| .30 to .50 (-.30 to -.50) | Low positive (negative) correlation |
| .00 to .30 (.00 to -.30) | Negligible correlation |

- A **positive** value means both variables increase together
- A **negative** value means as one increases, the other decreases
- A value close to **zero** means little to no correlation
- A value of **1** or **-1** means perfect linear correlation

## Section 2.1: Introduction to Data Maps (Discussion)

**Data maps** (map visualizations) allow us to visualize data directly on a map. These maps can help in identifying patterns and trends that might not be apparent in data tables.

### Key features of data maps:

**Overlaying Data onto Maps:** Data maps allow us to overlay various data points onto geographical maps, helping us visualize the distribution of data. This is very useful for our car accidents project!

**Heat Maps:** Heat maps are a type of data map where data values are represented as colors. This can help in identifying areas with higher concentrations of data points.

**Hover-over Details:** This feature allows users to hover over specific points on the map to view more detailed information about that point.
